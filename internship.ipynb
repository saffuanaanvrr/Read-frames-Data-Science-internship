```python
import tensorflow_hub as hub
import tensorflow as tf
import numpy as np
import cv2
import os

# Load the pre-trained model from TensorFlow Hub
MODEL_URL = "https://tfhub.dev/tensorflow/ssd_mobilenet_v2/2"
detector = hub.load(MODEL_URL)
def detect_objects(frame):
    input_tensor = tf.convert_to_tensor(frame)
    input_tensor = input_tensor[tf.newaxis, ...]
    detections = detector(input_tensor)
    return detections

def save_cropped_objects(frame, detections, output_folder, frame_count):
    height, width, _ = frame.shape
    bboxes = detections['detection_boxes'][0].numpy()
    scores = detections['detection_scores'][0].numpy()

    for i, bbox in enumerate(bboxes):
        if scores[i] > 0.5:
            ymin, xmin, ymax, xmax = bbox
            x1, y1, x2, y2 = int(xmin * width), int(ymin * height), int(xmax * width), int(ymax * height)
            cropped_object = frame[y1:y2, x1:x2]
            object_filename = os.path.join(output_folder, f"frame_{frame_count}_object_{i}.jpg")
            cv2.imwrite(object_filename, cv2.cvtColor(cropped_object, cv2.COLOR_RGB2BGR))


output_folder = 'cropped_objects'
os.makedirs(output_folder, exist_ok=True)


video_path = '/content/synthetic_video.mp4'
video_capture = cv2.VideoCapture(video_path)

frame_count = 0

while True:
    ret, frame = video_capture.read()
    if not ret:
        break

    frame_count += 1
    print(f'Processing frame {frame_count}')
    input_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    input_frame = input_frame.astype(np.uint8)
    detections = detect_objects(input_frame)


    save_cropped_objects(input_frame, detections, output_folder, frame_count)


video_capture.release()
print('Object detection and cropping completed.')
```
